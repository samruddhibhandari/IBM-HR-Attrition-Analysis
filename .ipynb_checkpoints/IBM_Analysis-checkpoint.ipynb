## Statistical Analysis

Before diving into individual factors, we establish a statistical 
baseline of the entire dataset to understand the shape, spread, 
and center of every numeric feature.

```python
# ============================================================
# STAT 1 — SKEWNESS CHECK ON ALL NUMERIC COLUMNS
# ============================================================
# Skewness tells us if data is balanced or pulled to one side
# Right skewed (>1) → few very high values pulling average up
# Use MEDIAN not mean to describe right skewed columns

skew_cols = ['MonthlyIncome', 'Age', 'YearsAtCompany',
             'YearsSinceLastPromotion', 'DistanceFromHome',
             'TotalWorkingYears', 'NumCompaniesWorked']

print("=== Skewness Check ===")
print(f"{'Column':<30} {'Skewness':>10}  {'Status':>15}")
print("-" * 58)
for col in skew_cols:
    skew = df[col].skew()
    flag = "⚠️ right skewed" if skew > 1 else "✅ normal"
    print(f"{col:<30} {skew:>10.2f}  {flag}")

print("\n📌 Right skewed columns → use MEDIAN not mean")
print("📌 These columns may need log transform before ML")

# Visualize skewness
plt.figure(figsize=(10, 5))
skew_values = df[skew_cols].skew().sort_values(ascending=False)
skew_values.plot(kind='bar', color='steelblue', edgecolor='white')
plt.axhline(y=1, color='red', linestyle='--', label='Right skew threshold (1.0)')
plt.axhline(y=-1, color='green', linestyle='--', label='Left skew threshold (-1.0)')
plt.title('Skewness of Key Numeric Features')
plt.ylabel('Skewness Score')
plt.legend()
plt.tight_layout()
plt.show()
# Columns above red line are right skewed
# MonthlyIncome is typically highly right skewed in HR data
```

#### Age- Impact of Age on Employee Attrition

```python
filt = df.groupby('Department')['Age'].mean().reset_index(name = 'AvgAge')
filt
```

```python
bins = [18, 25, 35, 45, 60]
labels = ['18-25', '26-35', '36-45', '46-60']
df['AgeBrackets'] = pd.cut(df['Age'], bins = bins,labels =labels)

attrition_rate = (df.groupby('AgeBrackets',observed=True)['Attrition'].mean()*100).reset_index(name='AttritionRate')
attrition_rate.round(2)
```

```python
plt.plot(attrition_rate['AgeBrackets'],attrition_rate['AttritionRate'],marker='o')
plt.title('Attrition Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Attrition Rate')
for i, v in enumerate(attrition_rate['AttritionRate']):
    plt.text(i, v, f"{v:.2f}%", ha='center')
plt.show()
```

#### Conclusion
Employee attrition is highest in the 18–25 age group, indicating that younger employees are more likely to leave the organization. While attrition generally decreases with age, there is a slight increase in the 46–60 group, suggesting some late-career transitions or retirements.

#### 📐 Statistical Analysis — Age
We check the central tendency, spread, and outliers in Age 
distribution before drawing conclusions about age and attrition.

```python
# ============================================================
# STAT 2 — AGE DISTRIBUTION STATISTICS + IQR OUTLIERS
# ============================================================
# Central tendency
age_mean   = df['Age'].mean()
age_median = df['Age'].median()
age_std    = df['Age'].std()
age_skew   = df['Age'].skew()

print("=== Age — Central Tendency & Spread ===")
print(f"Mean:     {age_mean:.2f}")
print(f"Median:   {age_median:.2f}")
print(f"Std Dev:  {age_std:.2f}")
print(f"Skewness: {age_skew:.2f}")

# IQR outlier detection
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df['Age'] < lower) | (df['Age'] > upper)]

print(f"\n=== Age — IQR Outlier Detection ===")
print(f"Q1: {Q1}  Q3: {Q3}  IQR: {IQR}")
print(f"Normal Range: {lower:.1f} to {upper:.1f}")
print(f"Outliers: {outliers.shape[0]}")

# Conditional probability — P(Attrition | AgeGroup)
print("\n=== P(Attrition | Age Group) ===")
overall = df['Attrition'].mean() * 100
bins   = [18, 25, 35, 45, 60]
labels = ['18-25', '26-35', '36-45', '46-60']
df['AgeBracket_stat'] = pd.cut(df['Age'], bins=bins, labels=labels)
age_prob = df.groupby('AgeBracket_stat', observed=True)['Attrition'].mean() * 100
for group, rate in age_prob.items():
    lift = rate - overall
    print(f"  {group}: {rate:.1f}%  ({lift:+.1f}% vs {overall:.1f}% average)")

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].hist(df['Age'], bins=20, color='steelblue', edgecolor='white')
axes[0].axvline(age_mean,   color='red',   linestyle='--', label=f'Mean: {age_mean:.1f}')
axes[0].axvline(age_median, color='green', linestyle='--', label=f'Median: {age_median:.1f}')
axes[0].set_title('Age Distribution')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')
axes[0].legend()
sns.boxplot(x=df['Age'], ax=axes[1], color='lightblue')
axes[1].set_title('Age Boxplot — IQR & Outliers')
axes[1].set_xlabel('Age')
plt.tight_layout()
plt.show()
# Box = middle 50% of employees (IQR)
# Dots outside whiskers = outliers
```

#### 📐 Conditional Probability — Job Factors
We calculate P(Attrition | Factor) for each job factor to measure 
exactly how much each factor increases attrition risk above baseline.

```python
# ============================================================
# STAT 4 — CONDITIONAL PROBABILITY FOR ALL KEY FACTORS
# ============================================================
# P(Attrition | X) = probability of leaving GIVEN condition X
# Lift = how much higher than overall average

overall = df['Attrition'].mean() * 100
print(f"Overall Attrition Rate: {overall:.2f}%\n")

factors = ['Department', 'OverTime', 'MaritalStatus',
           'JobRole', 'BusinessTravel', 'Gender']

for col in factors:
    print(f"=== P(Attrition | {col}) ===")
    prob = df.groupby(col)['Attrition'].mean() * 100
    for group, rate in prob.items():
        lift = rate - overall
        direction = "↑" if lift > 0 else "↓"
        print(f"  {str(group):<35} {rate:.1f}%  "
              f"{direction} {abs(lift):.1f}% vs average")
    print()
```

#### 📐 Correlation Analysis — Strongest Attrition Predictors
We measure correlation between every numeric feature and Attrition 
to identify which factors predict employee turnover most strongly.
These top features will guide our ML model's feature selection.

```python
# ============================================================
# STAT 5 — CORRELATION WITH ATTRITION (FIXED v2)
# ============================================================

# Step 1 — numeric columns only
numeric_df = df.select_dtypes(include='number')

# Step 2 — drop columns that are ALL NaN
numeric_df = numeric_df.dropna(axis=1, how='all')

# Step 3 — correlation with Attrition
corr = (numeric_df.corr()['Attrition']
                  .abs()
                  .sort_values(ascending=False)
                  .drop('Attrition')
                  .dropna())

print("=== Top 15 Features Correlated With Attrition ===")
print(f"{'Feature':<35} {'Correlation':>12}")
print("-" * 50)
for feat, val in corr.head(15).items():
    bar = '█' * int(val * 30)
    print(f"{feat:<35} {val:.4f}  {bar}")

print("\nInterpretation:")
print("0.00–0.10 → Negligible")
print("0.10–0.30 → Weak")
print("0.30–0.50 → Moderate")
print("0.50+     → Strong")

# Step 4 — heatmap — only use columns with valid correlations
top_features = corr.head(10).index.tolist() + ['Attrition']

# Keep only columns that exist in numeric_df
top_features = [f for f in top_features if f in numeric_df.columns]

# Build correlation matrix and drop any remaining NaN rows/cols
corr_matrix = numeric_df[top_features].corr()
corr_matrix = corr_matrix.dropna(axis=0, how='all')
corr_matrix = corr_matrix.dropna(axis=1, how='all')

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix,
            annot=True,
            fmt='.2f',
            cmap='coolwarm',
            center=0,
            linewidths=0.5)
plt.title('Correlation Heatmap — Top 10 Attrition Predictors')
plt.tight_layout()
plt.show()

# Bar chart
plt.figure(figsize=(10, 6))
corr.head(10).plot(kind='bar', color='coral', edgecolor='white')
plt.title('Top 10 Features Correlated With Attrition')
plt.ylabel('Absolute Correlation')
plt.xlabel('Feature')
plt.tight_layout()
plt.show()
```

## Statistical Summary

Before building our ML model, we consolidate all statistical 
findings to identify the strongest attrition signals in the dataset.
This directly guides our feature selection for ML.

```python
# ============================================================
# STAT 6 — COMPLETE STATISTICAL SUMMARY
# ============================================================

print("=" * 60)
print("   STATISTICAL SUMMARY — IBM HR ATTRITION ANALYSIS")
print("=" * 60)

overall = df['Attrition'].mean() * 100
print(f"\n📊 Overall Attrition Rate: {overall:.2f}%")
print(f"   ({int(df['Attrition'].sum())} out of {len(df)} employees left)\n")

# Highest risk groups
print("📊 Highest Risk Groups:")
print(f"{'Group':<35} {'Rate':>6}  {'vs Average':>12}")
print("-" * 57)

risk_groups = {
    'OverTime = Yes'          : df[df['OverTime'] == 'Yes'],
    'MaritalStatus = Single'  : df[df['MaritalStatus'] == 'Single'],
    'JobSatisfaction = 1'     : df[df['JobSatisfaction'] == 1],
    'BusinessTravel = Frequent': df[df['BusinessTravel'] == 'Travel_Frequently'],
    'JobLevel = 1 (Junior)'   : df[df['JobLevel'] == 1],
    'YearsAtCompany <= 2'     : df[df['YearsAtCompany'] <= 2]
}

for name, subset in risk_groups.items():
    rate = subset['Attrition'].mean() * 100
    lift = rate - overall
    print(f"  {name:<33} {rate:>5.1f}%  (+{lift:.1f}% above avg)")

# Top correlations
print(f"\n📊 Top 5 Features Correlated With Attrition:")
for i, (feat, val) in enumerate(corr.head(5).items()):
    print(f"  {i+1}. {feat:<35} r = {val:.4f}")

# Income comparison
left_median   = df[df['Attrition']==1]['MonthlyIncome'].median()
stayed_median = df[df['Attrition']==0]['MonthlyIncome'].median()
print(f"\n📊 Salary Gap:")
print(f"  Employees who left   → Median: ₹{left_median:.0f}")
print(f"  Employees who stayed → Median: ₹{stayed_median:.0f}")
print(f"  Gap: ₹{stayed_median - left_median:.0f} higher for those who stayed")

# Skewed features
print(f"\n📊 Right-Skewed Features (use median, not mean):")
skewed = df.select_dtypes(include='number').skew()
for feat, val in skewed[skewed > 1].sort_values(ascending=False).items():
    print(f"  {feat:<35} skewness = {val:.2f}")

print("\n" + "=" * 60)
print("  These findings directly guide ML feature selection.")
print("=" * 60)
```

## 6. Machine Learning — Attrition Prediction

We now build a model to predict which employees are likely to leave.
This transforms our EDA insights into an actionable prediction tool
for the HR team.

### Why Classification?
Attrition is a binary outcome — Yes or No.
This is a classification problem, not regression.

### Models We Build
1. Logistic Regression — simple, interpretable, good baseline
2. Decision Tree — visual, explainable to non-technical HR team

### Why Recall Over Accuracy?
Missing an employee who WILL leave (false negative) costs more
than flagging someone who won't (false positive).
HR would rather investigate 10 false alarms than miss 1 real case.
Therefore we optimize for **Recall**.

```python
# ============================================================
# ML BLOCK 1 — IMPORTS
# ============================================================
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (accuracy_score, precision_score,
                             recall_score, f1_score,
                             confusion_matrix, classification_report,
                             roc_auc_score)
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

print("All ML libraries imported successfully ✅")
```

```python
# ============================================================
# ML BLOCK 2 — PREPARE DATA FOR ML (FIXED)
# ============================================================

df_ml = df.copy()

# Step 1 — drop useless columns (same value for everyone)
useless_cols = ['EmployeeCount', 'StandardHours',
                'Over18', 'EmployeeNumber']

# Step 2 — drop label/category columns (string/category types)
label_cols = ['AgeBrackets', 'AgeBracket_stat',
              'DistanceCategory', 'IncomeCategory',
              'HikeCategory', 'SalaryBracket',
              'SalaryGrowthPotential', 'TenureCategory',
              'RetentionRisk']

# Step 3 — drop derived numeric columns
derived_cols = ['IncomePerYear', 'SatisfactionScore',
                'IsHighRisk', 'CareerExperienceScore',
                'IsPromotedRecently', 'ExperienceGap',
                'EmployeeScore', 'SalaryScore',
                'TenureScore', 'ProfileScore']

# Combine all columns to drop
all_drop = useless_cols + label_cols + derived_cols

# Drop only columns that actually exist
all_drop_existing = [col for col in all_drop if col in df_ml.columns]
df_ml = df_ml.drop(columns=all_drop_existing)

print(f"Columns dropped: {len(all_drop_existing)}")
print(f"Remaining columns: {df_ml.shape[1]}")
print(f"Remaining rows: {df_ml.shape[0]}")
print(f"\nNaN check: {df_ml.isnull().sum().sum()} total NaN values")
print(f"\nRemaining columns:")
print(df_ml.columns.tolist())
```

```python
# Find which column has NaN values
print(df_ml.isnull().sum().sort_values(ascending=False).head(10))
```

```python
# Check what Attrition looks like right now
print(df['Attrition'].value_counts())
print(f"dtype: {df['Attrition'].dtype}")
print(f"NaN count: {df['Attrition'].isnull().sum()}")
```

```python
# ============================================================
# ML BLOCK 3 — ENCODE CATEGORICAL COLUMNS
# ============================================================
# ML models only understand numbers
# Text columns must be converted to numbers

# Check which columns are still text
text_cols = df_ml.select_dtypes(include='object').columns.tolist()
print(f"Text columns to encode: {text_cols}")

# Encode all text columns using get_dummies
# drop_first=True removes one dummy to avoid multicollinearity
df_ml = pd.get_dummies(df_ml,
                        columns=text_cols,
                        drop_first=True)

print(f"\nAfter encoding:")
print(f"Shape: {df_ml.shape}")
print(f"All numeric: {df_ml.select_dtypes(include='object').shape[1] == 0}")
# Should print True — all columns are now numeric
```

```python
# ============================================================
# ML BLOCK 4 — TRAIN TEST SPLIT
# ============================================================

# Separate features (X) and target (y)
X = df_ml.drop('Attrition', axis=1)  # everything except Attrition
y = df_ml['Attrition']               # only Attrition column

print(f"Features (X): {X.shape}")
print(f"Target (y):   {y.shape}")
print(f"\nAttrition distribution:")
print(f"Stayed (0): {(y==0).sum()} ({(y==0).mean()*100:.1f}%)")
print(f"Left   (1): {(y==1).sum()} ({(y==1).mean()*100:.1f}%)")

# Split into train (80%) and test (20%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 20% for testing
    random_state=42,      # same split every run
    stratify=y            # keeps same attrition ratio in both sets
)

print(f"\nTraining set: {X_train.shape[0]} employees")
print(f"Testing set:  {X_test.shape[0]} employees")

# Scale numeric features
# Important for Logistic Regression
# Decision Tree doesn't need scaling but doesn't hurt
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

print("\nData scaled successfully ✅")
print("Ready for model training")
```

```python
# ============================================================
# ML BLOCK 5 — TRAIN BOTH MODELS
# ============================================================

# Model 1 — Logistic Regression
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train_scaled, y_train)
lr_pred = lr_model.predict(X_test_scaled)
print("Logistic Regression trained ✅")

# Model 2 — Decision Tree
dt_model = DecisionTreeClassifier(
    random_state=42,
    max_depth=5,        # limit depth to prevent overfitting
    min_samples_leaf=10 # each leaf needs at least 10 employees
)
dt_model.fit(X_train, y_train)  # Decision Tree doesn't need scaling
dt_pred = dt_model.predict(X_test)
print("Decision Tree trained ✅")
```

```python
# ============================================================
# ML BLOCK 6 — EVALUATE BOTH MODELS
# ============================================================

def evaluate_model(name, y_test, y_pred):
    print(f"\n{'='*50}")
    print(f"  {name}")
    print(f"{'='*50}")
    print(f"Accuracy:  {accuracy_score(y_test, y_pred)*100:.2f}%")
    print(f"Precision: {precision_score(y_test, y_pred)*100:.2f}%")
    print(f"Recall:    {recall_score(y_test, y_pred)*100:.2f}%")
    print(f"F1 Score:  {f1_score(y_test, y_pred)*100:.2f}%")
    print(f"ROC-AUC:   {roc_auc_score(y_test, y_pred)*100:.2f}%")
    print(f"\nClassification Report:")
    print(classification_report(y_test, y_pred,
                                target_names=['Stayed','Left']))

evaluate_model("Logistic Regression", y_test, lr_pred)
evaluate_model("Decision Tree",       y_test, dt_pred)
```

```python
# ============================================================
# ML BLOCK 7 — CONFUSION MATRIX VISUALIZATION
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, name, pred in zip(axes,
                           ['Logistic Regression', 'Decision Tree'],
                           [lr_pred, dt_pred]):
    cm = confusion_matrix(y_test, pred)
    sns.heatmap(cm, annot=True, fmt='d',
                cmap='Blues', ax=ax,
                xticklabels=['Stayed','Left'],
                yticklabels=['Stayed','Left'])
    ax.set_title(f'{name}\nConfusion Matrix')
    ax.set_ylabel('Actual')
    ax.set_xlabel('Predicted')

plt.tight_layout()
plt.show()

# Reading confusion matrix:
# Top-left     → True Negatives  (correctly predicted stayed)
# Bottom-right → True Positives  (correctly predicted left)
# Top-right    → False Positives (predicted left, actually stayed)
# Bottom-left  → False Negatives (predicted stayed, actually left) ← minimize this
```

```python
# ============================================================
# ML BLOCK 8 — FEATURE IMPORTANCE
# ============================================================

# Decision Tree feature importance
importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': dt_model.feature_importances_
}).sort_values('Importance', ascending=False).head(15)

plt.figure(figsize=(10, 6))
sns.barplot(data=importance, x='Importance', y='Feature', color='steelblue')
plt.title('Top 15 Most Important Features — Decision Tree')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.show()

print("Top 10 Features the model uses to predict attrition:")
print(importance.head(10).to_string(index=False))
# These features should match your EDA findings
# If they do → your EDA was correct ✅
# If they don't → interesting insight worth investigating
```

```python
# ============================================================
# ML BLOCK 9 — MODEL COMPARISON SUMMARY
# ============================================================

comparison = pd.DataFrame({
    'Model': ['Logistic Regression', 'Decision Tree'],
    'Accuracy':  [accuracy_score(y_test, lr_pred)*100,
                  accuracy_score(y_test, dt_pred)*100],
    'Precision': [precision_score(y_test, lr_pred)*100,
                  precision_score(y_test, dt_pred)*100],
    'Recall':    [recall_score(y_test, lr_pred)*100,
                  recall_score(y_test, dt_pred)*100],
    'F1':        [f1_score(y_test, lr_pred)*100,
                  f1_score(y_test, dt_pred)*100],
    'ROC-AUC':   [roc_auc_score(y_test, lr_pred)*100,
                  roc_auc_score(y_test, dt_pred)*100]
}).round(2)

print("=" * 65)
print("           MODEL COMPARISON SUMMARY")
print("=" * 65)
print(comparison.to_string(index=False))
print("=" * 65)

# Determine winner
lr_recall = recall_score(y_test, lr_pred)*100
dt_recall = recall_score(y_test, dt_pred)*100

print(f"\n📊 Best Model for HR Use Case:")
if lr_recall > dt_recall:
    print(f"→ Logistic Regression wins on Recall ({lr_recall:.1f}%)")
    print(f"→ Better at catching employees who will actually leave")
else:
    print(f"→ Decision Tree wins on Recall ({dt_recall:.1f}%)")
    print(f"→ Better at catching employees who will actually leave")

print("\n📌 Remember: We optimize for Recall, not Accuracy")
print("   Missing a flight risk employee costs more than a false alarm")
```
