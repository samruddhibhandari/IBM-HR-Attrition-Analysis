#  IBM HR Analytics — Employee Attrition Analysis

## Business Problem

Employee attrition — employees voluntarily leaving a company — is one 
of the most expensive problems in human resources.

Key facts about attrition cost:
- Replacing one employee costs 50%–200% of their annual salary
- IBM loses institutional knowledge every time a senior employee leaves
- High attrition damages team morale and reduces productivity

### Questions We Will Answer

Q1. Why are employees leaving?
    → Identify the key factors driving attrition

Q2. Who is most likely to leave next?
    → Build a model to predict individual attrition risk

Q3. What should we do about it?
    → Give actionable recommendations to reduce attrition


## Analysis Questions

### Demographics & Attrition
1. Which age group has the highest attrition rate?
2. Does gender affect attrition?
3. Does marital status influence likelihood of leaving?
4. Does distance from home affect attrition?

### Compensation & Attrition
5. Do employees who leave earn less than those who stay?
6. Which department has the lowest average salary?
7. Does salary hike percentage affect attrition?

### Job Factors & Attrition
8. Which department has the highest attrition rate?
9. Which job role loses the most employees?
10. Does overtime work drive attrition?

### Satisfaction & Attrition
11. Do dissatisfied employees leave more?
12. Which factors — salary, satisfaction, or promotion — 
    predict attrition most strongly?

### Experience & Growth
13. Do employees who haven't been promoted recently leave more?
14. Does number of companies worked before affect loyalty?
15. Are veteran employees (10+ years) more stable?

## ML Strategy

### Target Variable
Attrition → Binary classification problem
Yes = 1 (employee left)
No  = 0 (employee stayed)

### Why Binary Classification?
We are predicting one of two outcomes — left or stayed.
This is not a regression problem (no continuous number to predict)
and not a clustering problem (we have labeled data).

### Models We Will Build

Model 1 → Logistic Regression
   Why: Simple, interpretable, perfect for binary classification
        HR managers can understand "these 3 factors increase 
        attrition probability by X%" 
        Good baseline model to beat

Model 2 → Decision Tree
   Why: Visual and explainable — we can literally show HR 
        "if overtime=Yes AND salary<5000 AND satisfaction<=2 
        then 78% chance of leaving"
        Easy to present to non-technical stakeholders

### Why NOT complex models like Random Forest or XGBoost?
For an internship project — interpretability matters more than 
accuracy. A model HR can understand and act on is more valuable 
than a black-box model with 2% better accuracy.
We keep it simple and explainable on purpose.

### How We Will Evaluate
- Accuracy      → what % did we predict correctly overall?
- Precision     → of employees we flagged as leaving, how many actually left?
- Recall        → of employees who actually left, how many did we catch?
- Confusion Matrix → visual breakdown of correct/wrong predictions

### Why Recall Matters More Than Accuracy Here?
Missing an employee who IS going to leave (false negative) is 
more costly than incorrectly flagging someone who isn't (false positive).
HR would rather have a false alarm than miss a flight risk employee.
Therefore we optimize for Recall over raw Accuracy.

## Statistical Analysis

Before diving into individual factors, we establish a statistical 
baseline of the entire dataset to understand the shape, spread, 
and center of every numeric feature.

```python
# ============================================================
# STAT 1 — SKEWNESS CHECK ON ALL NUMERIC COLUMNS
# ============================================================
# Skewness tells us if data is balanced or pulled to one side
# Right skewed (>1) → few very high values pulling average up
# Use MEDIAN not mean to describe right skewed columns

skew_cols = ['MonthlyIncome', 'Age', 'YearsAtCompany',
             'YearsSinceLastPromotion', 'DistanceFromHome',
             'TotalWorkingYears', 'NumCompaniesWorked']

print("=== Skewness Check ===")
print(f"{'Column':<30} {'Skewness':>10}  {'Status':>15}")
print("-" * 58)
for col in skew_cols:
    skew = df[col].skew()
    flag = "⚠️ right skewed" if skew > 1 else "✅ normal"
    print(f"{col:<30} {skew:>10.2f}  {flag}")

print("\n📌 Right skewed columns → use MEDIAN not mean")
print("📌 These columns may need log transform before ML")

# Visualize skewness
plt.figure(figsize=(10, 5))
skew_values = df[skew_cols].skew().sort_values(ascending=False)
skew_values.plot(kind='bar', color='steelblue', edgecolor='white')
plt.axhline(y=1, color='red', linestyle='--', label='Right skew threshold (1.0)')
plt.axhline(y=-1, color='green', linestyle='--', label='Left skew threshold (-1.0)')
plt.title('Skewness of Key Numeric Features')
plt.ylabel('Skewness Score')
plt.legend()
plt.tight_layout()
plt.show()
# Columns above red line are right skewed
# MonthlyIncome is typically highly right skewed in HR data
```

```python
bins=[18,25,35,45,60]
labels=['Young','Mid','Senior','Late Career']
df['AgeGroup']=pd.cut(df['Age'], bins=bins,labels=labels)
df
# Young employees (18–25) show the highest attrition, indicating early-career employees leave more frequently.
# This suggests career exploration and lack of long-term commitment.
```

```python
# pd.cut is right inclusive
bins=[0,5, 15,float('inf')]
labels=['Near','Moderate','Far']
df['DistanceCategory'] =pd.cut(df['DistanceFromHome'],bins= bins, labels=labels)
df
# Employees living far from the workplace show higher attrition , indicating long commute increases likelihood of leaving.
# This suggests travel fatigue and work-life imbalance affect retention.
```

```python
result = df.groupby('DistanceCategory', observed =False)['Attrition'].apply(lambda x: (x ==1).mean()*100).reset_index()
print(result)
# Employees in the Far distance category have the highest attrition, indicating long commuting distance increases employee turnover.
# This suggests travel fatigue and poor work-life balance lead to higher attrition.
```

```python
df['EmployeeScore'] = ((df['JobSatisfaction'] *2) +
                       (df['PerformanceRating'] *3) +
                       (df['WorkLifeBalance'] *1))
result = df.sort_values(by='EmployeeScore', ascending=False)[
    ['EmployeeScore', 'Department']
].head(10).reset_index()

result
# Top-ranked employees have high EmployeeScore driven mainly by strong performance ratings and high satisfaction levels.
# This indicates that performance and satisfaction together define high-value employees.
```

```python
import numpy as np
conditions = [
    (df['Attrition']==1) | 
    ((df['JobSatisfaction']<=2) & (df['OverTime']=='Yes')),
    (df['YearsAtCompany'] > 5) & (df['JobSatisfaction'] >= 3)
]
choices = ['High','Low']
df['RetentionRisk'] = np.select(conditions,choices,default='Medium')
df
# Employees classified as High Risk show the highest attrition , indicating that low satisfaction, overtime, and existing attrition signals strongly predict employee turnover.
# This suggests these factors are critical drivers of employee exit.
```

```python
df['SalaryScore']= (df['IncomePerYear'] /df['IncomePerYear'].max())*40
df['SatisfactionScore'] =(df['JobSatisfaction']/4)*30
df['TenureScore'] = (df['YearsAtCompany'] /df['YearsAtCompany'].max())*30
df['ProfileScore']  = (df['SalaryScore'] + df['SatisfactionScore'] + df['TenureScore']).round(2)
result = df.sort_values(by='ProfileScore',ascending =False)[[
    'ProfileScore', 'Department', 'JobRole', 'MonthlyIncome', 'YearsAtCompany'
]].head(5)
result
# Top-ranked employees have high ProfileScore driven mainly by higher income and longer tenure.
# This indicates compensation and experience are major contributors to employee value.
```

```python
conditions = [
              (df['PerformanceRating'] == 4) & (df['YearsSinceLastPromotion'] >= 2),
              (df['PerformanceRating']== 3) & (df['YearsSinceLastPromotion']>= 3)
             ]
choices =['High','Medium']
df['SalaryGrowthPotential'] = np.select(conditions,choices,default='Low')
result = df.groupby('SalaryGrowthPotential')['IncomePerYear'].mean().round(2).reset_index()
result
# Employees in the High Salary Growth Potential group have strong performance ratings and have not been promoted recently.
# This indicates they are likely under-recognized and are prime candidates for salary increase or promotion
```

```python
sns.barplot(data = result,x='IncomePerYear', y='SalaryGrowthPotential')
plt.title('Salary Growth Potential by Income Per Year')
```

```python
order=['Low','Medium','High']
fig, ax= plt.subplots(1,2,figsize=(12,5))
sns.barplot(x='SalaryGrowthPotential',y='IncomePerYear', data=result,order=order,ax=ax[0])
ax[0].set_title('Average Monthly Income')

sns.boxplot(x='SalaryGrowthPotential',y='IncomePerYear', order=order,data=df,ax=ax[1])
ax[1].set_title('Income Distribution')
```

```python
filt = df.groupby('Department').agg(
    TotalEmployees = ('Department', 'count'),
    HighRiskEmployeeCount =('IsHighRisk','sum'),
    AverageProfileScore = ('ProfileScore', 'mean'),
    AverageSatisfactionScore = ('SatisfactionScore','mean'),
   MostCommonTenureCategory = ('TenureCategory',lambda x: x.mode().iloc[0])
).reset_index().round(2)
filt
# Departments with lower satisfaction scores tend to have higher high-risk employee counts.
# This indicates a strong relationship between employee satisfaction and attrition risk.
```

```python
sns.barplot(data=filt,x='TotalEmployees',y='HighRiskEmployeeCount')
plt.title('Total vs High-Risk Employees by Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```
